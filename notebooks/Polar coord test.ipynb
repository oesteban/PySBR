```python
import nibabel as nib
import numpy as np
import scipy.ndimage as si
import json

import os.path as op
import sys
sys.path.append( op.abspath( '../pysbr' ) )

import polar as p
```

```python
with open( '../data/template/simulated/landmarks.txt' ) as f:
    landmarks = json.load(f)
    
p1 = landmarks['spots1'][0]
p2 = landmarks['spots2'][0]

im = nib.load( '../data/template/simulated/segmentation.nii.gz' )
A = im.get_affine()[0:3,0:3]
O = im.get_affine()[0:3,3]

im_data = si.binary_erosion( im.get_data().astype( np.uint8 ) )
im_data = si.binary_opening( im_data.astype( np.uint8 ) )

labels, nb_labels = si.label( im_data )

nib.save( nib.Nifti1Image( labels, im.get_affine(), im.get_header() ), '../data/template/simulated/labels.nii.gz' )
```

```python
im2 = np.zeros_like( im.get_data() )
im2[labels==1] = 1
im_erosion = si.binary_erosion( im2 )
boundary = im2 - im_erosion
idxs = np.argwhere( boundary == 1 )
points = np.array( [ (np.dot( A, i )+O) for i in idxs.astype( np.float32 ) ] )
boundary[ tuple( np.dot( np.linalg.inv( A ),  (p2 - O) ).astype( np.uint8 ) ) ] = 2
nib.save( nib.Nifti1Image( boundary, im.get_affine(), im.get_header() ), '../data/template/simulated/boundary1.nii.gz' )
```

```python
X,Y,Z1 = p.polar_rep( points, p2 )
print Z1.mean()
print np.ma.median(Z1)
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import matplotlib.cm as cm
plt.imshow( Z1, cmap=cm.gray )
```

```python
plt.ion()
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
wframe = ax.plot_wireframe( X, Y, Z1, rstride=4, cstride=4 )
```

```python
im3 = np.zeros_like( im.get_data() )
im3[labels==2] = 1
im_erosion = si.binary_erosion( im3 )
boundary = im3 - im_erosion
boundary = boundary[::-1,:,:]
idxs = np.argwhere( boundary == 1 )
points = np.array( [ (np.dot( A, i )+O) for i in idxs.astype( np.float32 ) ] )

idx1 = np.dot( np.linalg.inv( A ),  (p1 - O) )
idx1[0] = np.shape(boundary)[0] - idx1[0] -1

newp1 = (np.dot( A, idx1 )+O)

boundary[ tuple( ( idx1 ).astype( np.uint8 ) ) ] = 2
nib.save( nib.Nifti1Image( boundary, im.get_affine(), im.get_header() ), '../data/template/simulated/boundary2.nii.gz' )
```

```python
X,Y,Z2 = p.polar_rep( points, newp1 )
print Z2.mean()
plt.imshow(Z2, cmap=cm.gray)

```

```python
tZ2 = si.affine_transform(Z2, np.identity(2) ,mode='wrap',offset=[-20,10] )
plt.imshow( tZ2, cmap=cm.gray )
```
